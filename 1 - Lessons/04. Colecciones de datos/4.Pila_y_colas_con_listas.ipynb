¬øA qu√© se refiere pila?

Es una colecci√≥n de elementos ordenados, **solo permite dos acciones: a√±adir un elemento a la pilo o sacarlo**.


Pila o 'stack' en estructura es lo com√∫nmente conocido en ingl√©s como 'LIFO', es decir 'Last In First Out'. Imaginemos que tenemos una monta√±a de libros, si queremos acceder al tercer libro, tendremos que quitar los dos primeros. Esos dos primeros libros llegaron a estar arriba del mont√≥n porque ser√≠an los √∫ltimos en colocarse en √©ste.

# COLAS

B√°sicamente el m√©todo contrario que las pilas, conocido en ingl√©s como 'FIFO', 'First In First Out'.

El primero que llega, el primero que se va (en un mundo ideal en donde el primero que llega es al primero al que atienden üòÜ).

```python
#Para trabajar con ello necesitaremos importar una librer√≠a
from collections import deque
```

```python
cola = deque()
```

```python
#Estar√° vac√≠a
cola
```

```python
cola = deque(['Primer paciente','Segundo paciente','Tercer paciente'])
```

```python
cola
```

```python
cola.append('Cuarto paciente')
```

```python
cola.append('Quinto paciente')
```

El ejemplo lo he hecho con pacientes en el que indico su orden de llegada a la consulta (primer, segundo,ect), para que quede m√°s claro que el paciente que llega primero es el que antes es atendido y por tanto antes se marcha (un ejemplo claro de un 'First In First Out').

Cabe **recordar** en las colas a diferencia de las pilas, **se saca el primero que ha llegado**, lo cual el pop es por el principio (horizontalmente la izquierda va primero, ya que escribimos de izquierda a derecha y los numero ordenados van de izquierda a derecha de menor a menor [siguiento un poco la l√≥gica]).
**Por tanto, usaremos un 'popleft'.**


```python
cola.popleft()
```
