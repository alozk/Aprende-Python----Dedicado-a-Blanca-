#PILA

¿A qué se refiere pila?

Es una colección de elementos ordenados, **solo permite dos acciones: añadir un elemento a la pilo o sacarlo**.


Pila o 'stack' en estructura es lo comúnmente conocido en inglés como 'LIFO', es decir 'Last In First Out'. Imaginemos que tenemos una montaña de libros, si queremos acceder al tercer libro, tendremos que quitar los dos primeros. Esos dos primeros libros llegaron a estar arriba del montón porque serían los últimos en colocarse en éste.




image.png

```python
pila = [5,6,7]
```

```python
pila.append(6)
```

```python
pila.append(3)
```

```python
#Los dos últimos valores añadidos aparecen al final
pila
```

```python
#Para extraerlos hay que hacerlo por el final como hemos comentado, con 'pop'
#Al usar 'pop', no solo nos muestra dicho valor sino que lo elimina tambien
pila.pop()
```

```python
pila
```

```python
#Es conveniente crear una variable para los 'pop' para que no perdamos dichos valores
n = pila.pop()
```

```python
n
```

```python
pila
```

```python
#Si hacemos 3 'pop' más, la pila se vacía pero siempre podríamos guardarlos con el método anterior de 'n' si hiciera falta
#Imprimirá el último valor extraído de la pila
pila.pop()
pila.pop()
pila.pop()
```

# COLAS

Básicamente el método contrario que las pilas, conocido en inglés como 'FIFO', 'First In First Out'.

El primero que llega, el primero que se va (en un mundo ideal en donde el primero que llega es al primero al que atienden 😆).

```python
#Para trabajar con ello necesitaremos importar una librería
from collections import deque
```

```python
cola = deque()
```

```python
#Estará vacía
cola
```

```python
cola = deque(['Primer paciente','Segundo paciente','Tercer paciente'])
```

```python
cola
```

```python
cola.append('Cuarto paciente')
```

```python
cola.append('Quinto paciente')
```

```python
cola
```

El ejemplo lo he hecho con pacientes en el que indico su orden de llegada a la consulta (primer, segundo,ect), para que quede más claro que el paciente que llega primero es el que antes es atendido y por tanto antes se marcha (un ejemplo claro de un 'First In First Out').

Cabe **recordar** en las colas a diferencia de las pilas, **se saca el primero que ha llegado**, lo cual el pop es por el principio (horizontalmente la izquierda va primero, ya que escribimos de izquierda a derecha y los numero ordenados van de izquierda a derecha de menor a menor [siguiento un poco la lógica]).
**Por tanto, usaremos un 'popleft'.**


```python
cola.popleft()
```

```python
cola.popleft()
```
